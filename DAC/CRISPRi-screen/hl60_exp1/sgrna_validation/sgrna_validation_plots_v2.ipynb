```r
#2020-02-01

#import packages
library('ggplot2')
library('RColorBrewer')
library('gridExtra')
```

```r
dcp2 <- read.table('dcp2_line_melt.txt', header = TRUE, sep = '\t')
dcp2$sgrna <- relevel(dcp2$sgrna, 'non-targeting sgRNA')

head(dcp2)
```

```r
p1 <- ggplot(dcp2, aes(x = time, y = mean, color = sgrna)) +
geom_line(size = 0.8) +
theme_classic() +
theme(plot.title = element_text(size = rel(1.75), color = 'black', face = 'bold', hjust = 0.5),
      axis.text.x = element_text(size = rel(1.75), color = 'black', angle = 0),
      axis.text.y = element_text(size = rel(1.75), color = 'black'),
      axis.title.x = element_text(size = rel(1.5)),
      axis.title.y = element_text(size = rel(1.5), margin = margin(r = 10)),
      legend.title = element_blank(),
      legend.text = element_text(size = rel(1), color = 'black'),
      legend.position = 'none'
      ) +
labs(title = 'DCP2', x = 'Day', y = 'Relative fraction of\nsgRNA+ cells') +
scale_x_continuous(limits = c(0,16), breaks = seq(0,15,5)) +
scale_y_continuous(limits = c(-0.05, 1.2), breaks = seq(0,1.2,0.2)) +
scale_color_manual(values = c('black', brewer.pal(4, 'Paired')[1:2])) +
geom_errorbar(aes(ymin = mean - sd, ymax = mean + sd, color = sgrna), width = 0.5) +
geom_point(size = 5)
```
