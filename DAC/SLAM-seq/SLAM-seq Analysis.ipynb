```r
#import tcounts files

fnames = list.files(path = '/avicenna/alex_ge/AGLG06/slamdunk_output/count', pattern = 'tcount.tsv', full.names = TRUE)

tcountfiles <- lapply(fnames, read.table, header = TRUE, sep = '\t', skip = 2)
```

```r
tcountfiles[[2]]
```

```r
tcountfiles_readscpm <- lapply(tcountfiles, function(x) x[x$ReadsCPM != 0, ])
```

```r
boxplot(tcountfiles_readscpm[[1]]$ConversionRate * 100, ylim = c(0.0, 4.0), outline = FALSE)
```

```r
tcountfiles_conversionrates <- do.call(cbind, lapply(tcountfiles_readscpm, function(x) x[, 7]))
```

```r
boxplot(tcountfiles_conversionrates * 100, ylim = c(0.0, 4.0), outline = FALSE)
```

```r
fnames_short <- lapply(fnames, substring, 48, 52)
colnames(tcountfiles_conversionrates) <- unlist(fnames_short)
```

```r
tcountfiles_conversionrates_df <- as.data.frame(tcountfiles_conversionrates)
```

```r
library(ggplot2)
library(reshape2)
tplot <- melt(tcountfiles_conversionrates_df)
```

```r
p <- ggplot(tplot, aes(x = variable, y = value)) + 
geom_boxplot(outlier.shape = NA) +
ylim(0,0.04) +
xlab('Condition') +
ylab('T>C Conversion Rate') +
theme_bw() +
theme(axis.text.x = element_text(size = rel(1.3), angle = 45, hjust = 1, color = 'black'),
      axis.text.y = element_text(size = rel(1.5), color = 'black'),
      axis.title.x = element_text(size = rel(1.5)),
      axis.title.y = element_text(size = rel(1.5)))
p

ggsave('conversion_rate_boxplot.pdf')
```
